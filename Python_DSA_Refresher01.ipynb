```python
 # Writing a program that asks the user for a number of days. The program then prints out the number of seconds in the amount of days given.
 # Constants
SECONDS_IN_A_MINUTE = 60
MINUTES_IN_AN_HOUR = 60
HOURS_IN_A_DAY = 24

# Create a function to calculate seconds in given days
def calculate_seconds_in_days(days):
    return days * HOURS_IN_A_DAY * MINUTES_IN_AN_HOUR * SECONDS_IN_A_MINUTE

# Using Try and Except to handle invalid input
def main():
    try:
        # Get user input
        days = int(input("Enter the number of days: "))

        # Calculate seconds using the function
        seconds = calculate_seconds_in_days(days)

        # Print the result to the user
        print(f"There are {seconds} seconds in {days} days.")
    except ValueError:
        print("Please enter a valid integer for the number of days.")
```
