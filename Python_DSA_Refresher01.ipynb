```python
 # Writing a program that asks the user for a number of days. The program then prints out the number of seconds in the amount of days given.
 # Constants Defined...
SECONDS_IN_A_MINUTE = 60
MINUTES_IN_AN_HOUR = 60
HOURS_IN_A_DAY = 24
# Create a function to calculate seconds in given days
def calculate_seconds_in_days(days):
    return days * HOURS_IN_A_DAY * MINUTES_IN_AN_HOUR * SECONDS_IN_A_MINUTE
# Using Try and Except to handle invalid input
def main():
    try:
        # Get user input
        days = int(input("Enter the number of days: "))
        # Calculate seconds by calling the function calculate_seconds_in_days
        seconds = calculate_seconds_in_days(days)
        # Print the result to the user
        print(f"There are {seconds} seconds in {days} days.")
    except ValueError:
        print("Please enter a valid integer words not accepted  for the number of days.")
 # Calling the  main function
main()
```

```python
#Writing a program that asks a user to input the radius then the program calculates the volume of a sphere (the formula for the volume is (4/3)Ï€r3). Use the exponential operator in python to compute (r3).
import math # Importing math module to access pi constant
# Function to calculate the volume of a sphere
def calculate_sphere_volume(radius):
    return (4/3) * math.pi * (radius ** 3)
# Main function to get user input and display the volume of the sphere
# Using Try and Except to handle invalid user input
def main():
    try:# try block to catch invalid input
        # Get user input for radius
        radius = float(input("Enter the radius of the sphere: "))
        # Calculate volume by calling the function calculate_sphere_volume
        volume = calculate_sphere_volume(radius)
        # Print the result to the user
        print(f"The volume of the sphere with radius {radius} is {volume:.2f}.")
    except ValueError:# except block to handle invalid input
        print("Please enter a valid number for the radius.")
# Calling the main function
main()
```

```python
#Using functions, writing a program to compute the area and perimeter of a square. The program should ask the user to enter a number corresponding to the side length of the square and display the area and perimeter of the square

# Function to calculate area of a square
def calculate_square_area(side_length):
    return side_length ** 2
# Function to calculate perimeter of a square
def calculate_square_perimeter(side_length):
    return 4 * side_length
# Main function to get user input and display area and perimeter of the square
# Using Try and Except to handle invalid user input
def main():
    try:# try block to catch invalid input
        # Get user input for side length
        side_length = float(input("Enter the side length of the square(Numerical Values Only): "))
        # Calculate area and perimeter by calling the respective functions
        area = calculate_square_area(side_length)
        perimeter = calculate_square_perimeter(side_length)
        # Print the results to the user
        print(f"The area of the square with side length {side_length} is {area:.2f}.")
        print(f"The perimeter of the square with side length {side_length} is {perimeter:.2f}.")
    except ValueError:# except block to handle invalid input
        print("Please enter a valid number for the side length.")
# Calling the main function
main()
```

```python
#Writing  a program using functions that determines whether a character input by a user is uppercase or lower case.
# Function to check if character is uppercase or lowercase
def check_character_case(character): # using if-elif-else statements
    if character.isupper():# Check if character is uppercase
        return "uppercase"
    elif character.islower(): # Check if character is lowercase
        return "lowercase"
    else:
        return "not an alphabetic character"
# Main function to get user input and display character case
def main():
    # Get user input for character
    character = input("Enter a single character: ")
    if len(character) != 1: # Check if input is a single character
        print("Please enter exactly one character.")
        return
    # Check character case by calling the function check_character_case
    case = check_character_case(character)
    # Print the result to the user
    print(f"The character '{character}' is {case}.")
# Calling the main function
main()
```

```python
#Write a Python program that does the following:
#Uses a loop for a user to continually input 5 values to populate an array.
#Calculates and displays the average of the values input into the array.
# Function to get user input and calculate average
def main():
    numeric_values = [] # Initialize an empty list to store values
    # Loop to get 5 values from the user
    for i in range(5): # for loop to iterate 5 times
        while True: # Loop until valid input is received
            try: # Try block to catch invalid input
                value = float(input(f"Enter value {i + 1}: ")) # Get user input
                numeric_values.append(value) # Append the value to the list
                break # Break the loop if input is valid
            except ValueError:
                print("Please enter a valid number.") # Prompt for valid input
    # Calculate average
    average = sum(numeric_values) / len(numeric_values)
    # Print the average to the user
    print(f"The average of the entered values is: {average:.2f}") # Display average with 2 decimal places
# Calling the main function
main()
##
```
