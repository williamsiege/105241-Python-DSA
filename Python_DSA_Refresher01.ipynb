```python
 # Writing a program that asks the user for a number of days. The program then prints out the number of seconds in the amount of days given.
 # Constants Defined...
SECONDS_IN_A_MINUTE = 60
MINUTES_IN_AN_HOUR = 60
HOURS_IN_A_DAY = 24
# Create a function to calculate seconds in given days
def calculate_seconds_in_days(days):
    return days * HOURS_IN_A_DAY * MINUTES_IN_AN_HOUR * SECONDS_IN_A_MINUTE
# Using Try and Except to handle invalid input
def main():
    try:
        # Get user input
        days = int(input("Enter the number of days: "))
        # Calculate seconds by calling the function calculate_seconds_in_days
        seconds = calculate_seconds_in_days(days)
        # Print the result to the user
        print(f"There are {seconds} seconds in {days} days.")
    except ValueError:
        print("Please enter a valid integer words not accepted  for the number of days.")
 # Calling the  main function
main()
```

```python
#Write a program that asks a user to input the radius then the program calculates the volume of a sphere (the formula for the volume is (4/3)Ï€r3). Use the exponential operator in python to compute (r3).
import math
# Function to calculate the volume of a sphere
def calculate_sphere_volume(radius):
    return (4/3) * math.pi * (radius ** 3)
# Main function to get user input and display the volume of the sphere
# Using Try and Except to handle invalid user input
def main():
    try:# try block to catch invalid input
        # Get user input for radius
        radius = float(input("Enter the radius of the sphere: "))
        # Calculate volume by calling the function calculate_sphere_volume
        volume = calculate_sphere_volume(radius)
        # Print the result to the user
        print(f"The volume of the sphere with radius {radius} is {volume:.2f}.")
    except ValueError:# except block to handle invalid input
        print("Please enter a valid number for the radius.")
# Calling the main function
main()
```
